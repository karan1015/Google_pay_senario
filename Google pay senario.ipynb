```python
class Gpay(wish):
    def __init__(self,c_name,c_phone_no,c_pin,c_balance):
        self.name=c_name
        self.phone_no=c_phone_no
        self.pin=c_pin
        self.balance=c_balance
        
    def Access_object(self):
        print(self.name)
        print(self.phone_no)
        print(self.balance)
        print()
        self.customer()

    def start(self):
        pin=int(input("Enter Your Pin: "))
        if pin==self.pin:
            self.Access_object()
        else:
            print("\n INCORRECT PIN")
            
    #balance check
    def acc_balance(self):
        print("Balance: ",self.balance)
        choose=int(input("\ncontinue transaction press 1 exit press 0: "))
        if choose==1:
            self.customer()
        else:
            print("\n-----------THANK YOU---------------------")


    def send(self):
        amount=int(input("\nEnter your send Amount: "))
        if amount<=self.balance:
            self.balance=self.sub(self.balance,amount)
            print("\nAmount succesfully Send..")
            print("Balance: ",self.balance)
            choose=int(input("\ncontinue transaction press 1 exit press 0: "))
            if choose==1:
                self.customer()
            else:
                print("\n-----------THANK YOU---------------------")
        else:
            print("\nInsufficient Balance")
                    
    @staticmethod            
    def sub(num1,num2): 
        return num1-num2
    
    def M_send(self):
        amount=int(input("\nEnter Recharge Amount: "))
        if amount<=self.balance:
            self.balance=self.sub(self.balance,amount)
            print("\nRecharged Succesgully..")
            print("Balance: ",self.balance)
            choose=int(input("\ncontinue transaction press 1 exit press 0: "))
            if choose==1:
                self.customer()
            else:
                print("\n-----------THANK YOU---------------------")
        else:
            print("\nInsufficient Balance")
                    
    @staticmethod            
    def sub(num1,num2): 
        return num1-num2
            
#upi transaction        
class Upi_id():
    def Upi_trans(self):
        upi=input("Enter your Upi_id: ")
        flag=False
        for i in upi:
            if i in '@':
                flag=True
        if flag==True:
            self.send()
        else:
            print("\n INVALID UPI_ID")
#Mobile Recharge
class Mobile_recharge():
    def M_recharge(self):
        Mobile_no=input("Enter Mobile_No: ")
        length=len(Mobile_no)
        if length==10:
            print("\n 1.TOP-UP  2.DATA ADD-ON  3.POPULAR PLANS : ")
            plan=int(input("Enter Your Plan : "))
            if plan==1:
                self.M_send() 
            elif plan==2:
                self.M_send()
            elif plan==3:
                self.M_send()
            else:
                print("\nINVALID CHOOSE")
        else:
            print("\nINVALID MOBILE NUMBER")

#Account Transaction        
class account_trans(Gpay):
    def Acc_trans(self):
        acc_no=input("Enter Account_no : ")
        length=len(acc_no)
        if length==16:
            self.send()
        else:
            print("\nCheck Your Account_No")

class wish(Upi_id,Mobile_recharge,account_trans):
    def customer(self):
        print("\n1.Upi_Transcation 2.Mobile_Recharge  3.Balance_check  4.Account_transaction ")
        choose=int(input("\n\nEnter Your Option: "))
        if choose==1:
            self.Upi_trans()
        elif choose==2:
            self.M_recharge()
        elif choose==3:
            self.acc_balance()
        elif choose==4:
            self.Acc_trans()
        else:
            print("\n Invalid choose")
            
user=Gpay('karan',9080660113,1015,4500)
user.start()
```
